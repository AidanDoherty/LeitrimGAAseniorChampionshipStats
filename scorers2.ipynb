```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import os
data = pd.read_csv("scores2.csv") 
data = data.fillna(0)


```

```python
def getScoresByClub(club): 
    club = club.sort_values('Total')
    y_pos = np.arange(len(club['Name']))
    plt.bar(y_pos, club['Total'], align='edge', alpha=0.5)
    plt.xticks(y_pos, club['Name'],rotation=45)
    plt.ylabel('Points')
    plt.title(club['Team'].iloc[0] + ' - Total Points Scored (Points + Goals)')
    
    plt.show()


    
def getAllScores(data):
    scores = data.filter(['Name','Total'], axis=1)
    
    scores =  scores.sort_values('Total')
    scores.plot.barh(x="Name")
    
def teamScoresFromFrees(data):
    scores = data.filter(['Name','Frees'], axis=1)
    scores =scores[scores.Frees != 0]
    scores =  scores.sort_values('Frees')
    scores.plot.barh(x="Name")
    
def teamScoresFromPlay(data):
    club = data
    y_pos = np.arange(len(club['Name']))
    plt.barh(y_pos, club['Total'] -club['Frees'], align='edge', alpha=0.5)
    plt.xticks(y_pos, club['Name'],rotation=45)
    plt.ylabel('Points')
    plt.title(club['Team'].iloc[0] + ' - Total Points Scored (Points + Goals)')
    
    plt.show()

    
def playerGoals(data):
    scores = data.filter(['Name','Goals '], axis=1)
    scores =scores[scores['Goals ']!= 0]
    scores =  scores.sort_values('Goals ')
    scores.plot.barh(x="Name")

```
